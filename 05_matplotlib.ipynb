## Import matplotlib | Importando matplotlib

One of the easiest ways to make use of this library is through the `pyplot` module, into which we usually import it under the alias` plt`.

We will also be using numpy to initially create a dataset for visualization purposes.

---

Una de las formas más sencillas de hacer uso de esta librería es a través del módulo `pyplot`, al cual lo solemos importar bajo el alias `plt`.

Vamos a usar numpy para crear un set de datos que vamos a visualizar.

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Plot functions | Graficando de funciones

Many times we want to graphically visualize a mathematical function that depends on a certain variable.
For example, let's graph the sine function between 0 and 2 $\pi$. For this, we have to create an array of the values we will evaluate our function for (x). 

---

Muchas veces deseamos visualizar gráficamente una función matemática que depende de cierta variable.
Por ejemplo, grafiquemos la función seno entre 0 y 2 $\pi$. Para esto, debemos creear un arreglo con las valores que serán evaluadas en la función (x).

````python
x = np.linspace(0, 2 * np.pi, 100) # start point, end point, number of points | 0 to 2pi, 100 points
print(x)
````

Now we evaluate the sine function using all the values in x | Ahora evaluamos la función seno utilizando todos los valores en x 

````python
y = np.sin(x) 
````

Let's try to plot the `sin(x)` function | Intentemos plotear la función `sin(x)`:

````python
plt.plot(x, y) # x from 0 to 2pi, y = sin(x)
plt.show()
````

Now let's add the name of the axes and a figure title | Ahora agreguemos el nombre de los ejes y un título para la figura: 

````python
plt.plot(x, y)

# Add the name of the axes | Agrega los nombres de los ejes
plt.xlabel('Angle')
plt.ylabel('Sine')

# Add a title | Agrega un título
plt.title('Sine function')

plt.show()
````

## Multiple curves in the same plot | Múltiples curvas en un mismo gráfico

How can we include several curves in the same graph? | ¿Cómo podemos incluir varias curvas en un mismo gráfico?

Add the `label`, `linestyle` and `color`options to differentiate each of the curves. | Agregar las opciones `label`, `linestyle` y `color` para diferenciar cada un de las curvas:

We need to use the `plt.legend` function to see the curves names (`label`) | Necesitamos usar la función `plt.legend` para ver los nombres de las curvas (`label`).

````python
# 1st curve | Primera curva
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)", color="tab:blue")
# 2nd curve | Segunda curva
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)", color="tab:orange")
# 3th curve | Tercera curva
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)", color="tab:green")

# Add the legend | Agrega la leyenda
plt.legend()

# Add the name of x axes | Agrega el nombre del eje x
plt.xlabel('Angle')
# Add the name of y axes | Agrega el nombre del eje y 
plt.ylabel('Sine')
# Add a title | Agrega un título
plt.title('Sine function')

plt.show()
````

### Creating Subplots

What if you want to plot each curve in three different curves? | ¿Qué sucede si quiero graficar cada curva en un gráfico distinto?

We can use the matplotlib.pyplot `subplot` function | Podemos utilizar la función `subplot` de matplotlib.pyplot

````python
# Create a new figure with a specific size (width=15, height=4)
plt.figure(figsize=(15, 4))

# First plot | Primer gráfico
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st plot | 1 fila, 3 columnas, 1er gráfico
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)", color="tab:blue")
plt.title('sin(x)')

# Second plot | Segundo gráfico
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd plot | 1 fila, 3 columnas, 2do gráfico
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)", color="tab:orange")
plt.title('sin(2x)')

# Third plot | Tercer gráfico
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd plot | 1 fila, 3 columnas, 3er gráfico
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)", color="tab:green")
plt.title('sin(3x)')

#plt.legend()
plt.show()
````

## Good practices | Buenas prácticas

When we want to plot and publish data, we always have to make sure that our figures have: | A la hora de visualizar datos y publicarlos, siempre tenemos que asegurarnos de que nuestras figuras cuenten con:

- Annotations of the coordinate axes | Anotaciones de los ejes coordenados
- Axis units | Unidades de los ejes
- Legend (if we have several plots in the same figure) | Leyenda (si tenemos varios plots en la misma figura)

In addition, we can add information and some details such as: | Además, podemos añadir información y algunos detalles como:

- A title | Un título
- Decorations, like a grid | Decoraciones como un grid

Let's try the `grid` decoration | Probemos la decorración `grid`:

````python
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)")
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)")
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)")

# Add the legend | Agrega la leyenda
plt.legend()

# Add the name of x axes | Agrega el nombre del eje x
plt.xlabel("Angle")
plt.ylabel("y")

# Add the decoration | Agrega la decoracion
plt.grid()

plt.show()
````

## Plot data | Graficando datos

Now we will plot the mean temperature data for entire planet, from 1850 to the present.
It is available by [Berkeley Earth](http://berkeleyearth.org/).
The data files can be downloaded from their website, but in the repository we already have a simplified version of one of them in `data/temperature-berkeley.dat`.

We are going to import the `Path` class from` pathlib` which helps us to create paths to files and folders on our disk.
Let's define the path to the temperatures:
* Using `Path` to this, we ensure that this code can work on various operating systems.
  No matter what type of directory separator they use (`/` or `\`).
* Remember that the symbol `..` refers to the parent directory.

---

Vamos a graficar los datos de temperaturas medias anual de toda el planeta, desde 1850 hasta la actualidad.
Estos datos estan a disponibilidad por [Berkeley Earth](http://berkeleyearth.org/).
Los archivos de datos pueden ser descargados desde su página web, pero en el repositorio ya contamos con una versión simplificada de uno de ellos en `data/temperature-berkeley.dat`.

Vamos a importar la clase `Path` de `pathlib` que nos sirve para crear rutas a archivos y carpetas de nuestro disco.
Definamos la ruta al archivo de temperaturas:
* Al utilizar `Path` para esto, nos aseguramos que este código pueda funcionar en diversos sistemas operativos, sin importar qué tipo de separador de directorios utilizan (`/` o `\`) entre otras diferencias.
* Recordemos que el símbolo `..` hace referencia al directorio padre.


````python
from pathlib import Path
````

We create a variable called `berkeley_data_file` where we store the direction to the temperature file.
This variable is a _string_ type.

---

Creamos una variable llamada `berkeley_data_file` donde guardamos la dirección al archivo de temperaturas.
Esta variable es del tipo _string_.

````python
berkeley_data_file = Path(".")/ "data" / "temperature-berkeley.dat"

berkeley_data_file
````

Let's use the `numpy.loadtxt` function to read the file | Usemos la función `numpy.loadtxt` para leer el archivo.

````python
datos = np.loadtxt(berkeley_data_file)

datos
````

We can extract the columns corresponding years and temperatures from the `data` array | Podemos extraer las columnas correspondiente años y temperaturas del array `datos`

````python
years, temperaturas = datos[:, 0], datos[:, 1]
````

````python
plt.plot(years, temperaturas, "o-")

plt.show()
````

We can save the plot as a `png` image or `PDF` (vectorial image) | Podemos guardar el gráfico como una imagen `png` o` PDF` (imagen vectorial)

````python
plt.plot(years, temperaturas, "o")

# Add the axes names | Agrega el nombre de los ejes
plt.xlabel("Años")
plt.ylabel("Temperatura [C]")

# Add the plot title | Agrega el titulo de la gráfica
plt.title("Temperatura media anual de la Tierra")

# Add the decoration | Agrega la decoracion
plt.grid()

# Save the plot as png and pdf | Guarda la gráfica como png o pdf
plt.savefig("temperatura2.png", dpi=300)
plt.savefig("temperatura.pdf")

plt.show()
````

### pcolormesh

````python
# Data to plot.
x_aux = np.linspace(-np.pi, np.pi, 100)

# Create a grid | Creo una grilla
x, y = np.meshgrid(x_aux, x_aux)

# Calculate the values | Calculo los valores
z = np.sinc(x**2 + y**2)
````

````python
# Plot
plt.pcolormesh(x, y , z, shading="auto", cmap='RdBu', vmin=-1, vmax=1)

plt.colorbar()
````
